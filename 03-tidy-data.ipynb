```python
billboard = pd.read_csv('./data/billboard.csv')
```

```python
# we actually have 81 columns
print(billboard.shape)


#  example where we want to keep the first set of columns fixed while melting the rest of the columns
# I'm holding the year, artist, track, time, and date entered columns fixed
#  rest of the columns will be pivoted into a new variable called week
# And the values of those columns will be pivoted into a new column called rating
billboard.melt(id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
               value_name='rank',
               var_name='week')

# All of the week values have been melted into a single column, followed by a rating.
# This is a useful feature, because if you remember from the plotting exercises, 
# we can then do something like a faceted plot along the week column. 
# This is one of the reasons why tidy data is so important, 
# because it allows us to manipulate and visualize our data sets extremely quickly.

```

```python
(
billboard
  .melt(id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
        value_name='rank',
        var_name='week')
  .groupby('artist')['rank']
  .mean()
)
```

```python
# Here are the first five rows and twelve columns of a weather data set. You can see we have an ID, a year, a month, an element that tells us whether this is the max temperature or min temperature, and days.
# What we wanna do is melt down the days so that we have a column that represents day of the month, and another column for a value. 
# However, what's problematic is this element column, what we really want is to have a column that represents the day of the month, and the value for the max temperature and a value for the minimum temperature. 
weather = pd.read_csv('./data/weather.csv')
```

```python
weather.head()
```

```python
# So this is the problem that we're going to address here. Since I already know how to solve part of the problem, that is how do we take the days of the month and bring it down, we can use the melt function, so let's first do that.
weather_long = weather.melt(id_vars=['id', 'year', 'month', 'element'],
                            var_name='day',
                            value_name='temp')
```

```python
weather_long.head()
```

```python
# After we melt the data, you can see that we're half of the way there. 
# We have a column that represents the day of the month, and another column that represents the temperature. 
# The next problem we have is essentially how do we undo the melt command. The opposite of the melt command is using pivot table. 
# What pivot table does, is it takes a column, and takes each individual value of that column and turns it into another column.
# You then specify another column, here temperature, that will be used to fill in the rest of the table. 
# Here's the documentation for the pandas.pivot_table function one of the first things we need to specify is the index. 
# Again, those are the columns that will remain the same during this operation. 
# The next parameter that we need to specify is the columns, that's the column where we want to undo the melt command. 
# Lastly, we provide the values column that tells the pivot table function what are the values to fill in the cell. 

(weather_long.pivot_table(index=['id', 'year', 'month', 'day'],
                         columns='element',
                         values='temp')
.reset_index()
)
#  We can now see, after we reset the index, we now get a flat rectangular data set in the shape that we want. 
# So when you have a data set where variables are stored in both rows and columns, you first have to do a melt on a subset of the columns, and then do a pivot on one of the other columns. Finally, to make sure you get a nice square or rectangular data set, you have to reset all the indices.
```

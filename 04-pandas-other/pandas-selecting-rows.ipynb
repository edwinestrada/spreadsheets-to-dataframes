```python
# read csv and store result in variable
df = pd.read_csv("data\stock_data_simple.csv") 
```

```python
# print out dataframe 
df
```

```python
# sort by annual sales and save dataframe in variable
df = df.sort_values("annual_sales", ascending=False)
df
```

# Selecting Rows (Multiple Ways)
### a. Select the first 3 rows of a DataFrame

```python
# since rows numbers start at 0 in Pandas, 0 is first row
df.iloc[0:3,:]
```

### b. Select between row 4 and row 7

```python
# note that to include row 7, need to specify 8 as last row
df.iloc[4:8,:]
```

## c. Select all rows after a row 

```python
# since there are 10 rows in CSV File, we get the last 6
df.iloc[4: ,:]
```

## d. Select a Specific Row (fyi... there are multiple ways)

Say we wanted to select the the row for Microsoft

```python
# we could use the ticker column
df
```

```python
df['ticker'] == "MSFT"
```

```python
# if we wrap the statment above inside the original dataframe using square brackets
df[df['ticker'] == "MSFT"]
```

```python
# or can use isin
df['ticker'].isin(["MSFT"])
```

```python
# again, wrap in square brackets
df[df['ticker'].isin(["MSFT"])]
```

```python
# we could store the row in a new variable'
df_msft = df[df['ticker'].isin(["MSFT"])]
```

```python
# we can also get the row number
df_msft.index[0]
```

```python
# or we can store the row number one line
row_number = df[df['ticker'] == "MSFT"].index[0]
```

```python
print(row_number)
```
